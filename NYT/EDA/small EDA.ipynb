```python
import pickle

with open('NYTcorpus_train.p', mode='rb') as f:
    articles = pickle.load(f)

articles = [a[2] for a in articles]
```

```python
import re

pattern = '([A-Z]{3,}(\s|$))+'

def match(text):
    res = re.search(pattern, text)
    if res:
        return res[0]
    

keywords_in_end = [match(a[-100:]) for a in articles]

from collections import Counter

Counter(keywords_in_end).most_common(100)
```
